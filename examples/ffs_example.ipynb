```python
import os
import numpy as np

from astropy.io import fits
from pyaraucaria.ffs import FFS

fits_path = "~/mglab/data/fits/zb08c_1052_65023.fits"
```

```python
hdul = fits.open(fits_path)
data = hdul[0].data
hdr = hdul[0].header
hdul.close()
```

```python
# example 1 (simple)

ffs = FFS(data)
ffs.saturation = 50000
ffs.calc_frame_fwhm(threshold=10, fwhm=10, box=10, N_stars=20)

print(ffs.frame_fwhm)

# for k in ffs.stats.keys():
#     print(f'{k}: {ffs.stats[k]} // {ffs.stats_description[k]}')
```

```python
ffs = FFS(data)
ffs.saturation = 50000
ffs.mk_stats()
ffs.sky_gradient(n_segments=10)
ffs.find_lines()
ffs.find_stars(threshold=10,fwhm=3)
ffs.star_info(box=15,N_stars=20)
ffs.calc_star_stats()

for k in ffs.stats.keys():
    print(f'{k}: {ffs.stats[k]} // {ffs.stats_description[k]}')
```
